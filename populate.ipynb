```python
player_columns = ['bill_id','billing_month','units','bill_amount','due_date']

df = pd.DataFrame(columns=player_columns)
df
```

```python
from datetime import datetime, timedelta

for i in range(1,501):
    # Generate a random month and year
    random_month = faker.random_int(min=1, max=12)
    random_year = faker.random_int(min=2010, max=datetime.now().year)  # Restrict to current year

    # Set the initial date as the first day of the generated month
    initial_date = datetime(random_year, random_month, 1)

    # Calculate due date 10 days later
    due_date = initial_date + timedelta(days=10)

    # Add the row to the DataFrame
    df = df.append({
        'bill_id': i,
        'billing_month': initial_date.strftime('%B %Y'),
        'units': faker.random_int(min=1, max=100),
        'bill_amount': faker.random_int(min=10000, max=50000),
        'due_date': due_date.strftime('%d-%m-%Y')
    },ignore_index=True)
```

```python
#write dataframe to sheet called electricity_bill in csv

df.to_excel(writer,sheet_name='electricity_bill',index=False)
```

```python
student_df = pd.DataFrame(columns=[
    'studentERP', 'name', 'email', 'departmentID', 'marital_status', 'phone_number',
    'age', 'address', 'gender', 'date_joined', 'graduating_year', 'degree_ID', 'degree_status'
])

email = ['gmail.com','yahoo.com','hotmail.com','outlook.com']


for i in range(22000,23001):
    student_df = student_df.append({
        'studentERP': i,
        'name': faker.name(),
        'departmentID': faker.random_int(min=1, max=10),
        'marital_status': faker.random_element(elements=['Single', 'Married']),
        'phone_number': '0300' + str(faker.random_int(min=1000000, max=9999999)),
        'age': faker.random_int(min=18, max=30),
        'address': faker.address(),
        'gender' : faker.random_element(elements=['Male','Female']),
        'date_joined': faker.date_this_century().strftime('%d-%m-%Y'),
        'degree_ID': faker.random_int(min=1, max=13),
        'degree_status': faker.random_element(elements=['Enrolled', 'Graduated'])
    },ignore_index=True)

for i in range(22000,23001):
    student_df['email'] = student_df['name'].str.replace(' ', '_').str.lower() + '@' + faker.random_element(elements=email)
```
